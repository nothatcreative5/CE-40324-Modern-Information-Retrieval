    در این قسمت تابع clean_data() را پیاده سازی کنید. عملکرد تابع به این صورت است که یک متن به عنوان ورودی گرفته و توکن‌های ولید آن را به صورت یک لیست که عملیات lemmatization, stemming و case folding اجرا شده است خروجی می‌دهد. متن ورودی شامل عنوان مقاله یا چکیده آنهاست.

```python
import nltk
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from nltk.stem import WordNetLemmatizer
import string

def clean_data(text):
    """Preprocesses the text with tokenization, case folding, stemming and lemmatization, and punctuations

    Parameters
    ----------
    text : str
        The title or abstract of an article

    Returns
    -------
    list
        A list of tokens
    """
    tokens = word_tokenize(text.lower())

    # remove punctuation
    tokens = [token for token in tokens if token not in string.punctuation]
    
    stemmer = PorterStemmer()
    lemmatizer = WordNetLemmatizer()
    # tokens = [stemmer.stem(lemmatizer.lemmatize(token.lower())) for token in tokens]
    tokens = [lemmatizer.lemmatize(token.lower()) for token in tokens]
    
    return tokens


clean_data("An abstract is a summary of the main article.") # return ["an", "abstract", "is", "a", "summary", "of", "the", "main", "article"]
```

```python
## 2 points
import pandas as pd


def find_stop_words(num_token=30):
    """Detects stop-words

     Parameters
    ----------
    all_text : list of all tokens
        (result of clean_data(text) for all the text)

    Returns
    -------
    Return Value is optional but must print the stop words and number of their occurence
    """

    df = pd.read_csv("data.csv")

    df["abstract"] = df["abstract"].fillna("")

    all_text = ""

    for _, row in df.iterrows():
        all_text += row["title"] + " " + row["abstract"]
    
    all_text = clean_data(all_text)


    word_counts = {}
    for word in all_text:
        if word not in word_counts:
            word_counts[word] = 1
        else:
            word_counts[word] += 1

    sorted_counts = sorted(word_counts.items(), key=lambda x: x[1], reverse=True)

    stop_words = {}
    for word, count in sorted_counts[:num_token]:
        stop_words[word] = count

    return stop_words

    
stop_words = find_stop_words(5)
```

```python
## 12 points

def construct_positional_indexes(corpus : str):
    
    """
    Get processed data and insert words in that into a trie and construct postional_index and posting lists after wards.

    Parameters
    ----------
    corpus: str
        processed data 
    
    Return
    ----------
    docs: 
        list of docs with specificied id, title, abstract.
    """
    positional_index = {}


    '''
        Save the length of each part of the document in the positional index
    '''
    
    for i, row in df.iterrows():
        paper_id = i + 1
        title = row['title']
        abstract = row['abstract']


        '''
            0 -> title
            1 -> abstract

            token = [
            title -> [paper_id, length, [positions]],
            abstract -> [paper_id, length, [positions]]]
            
        '''
        
        # Create positional index for title tokens
        for i, token in enumerate(title):
            if token not in positional_index:
                positional_index[token] = []
            if len(positional_index[token]) == 0 or positional_index[token][-1][0] != paper_id:
                positional_index[token].append([paper_id,[len(title)],[len(abstract)]])
            title_list = positional_index[token][-1][1]
            title_list.insert(-1, i)


        for i, token in enumerate(abstract):
            if token not in positional_index:
                positional_index[token] = []
            if len(positional_index[token]) == 0 or positional_index[token][-1][0] != paper_id:
                positional_index[token].append([paper_id,[len(title)],[len(abstract)]])
            abstract_list = positional_index[token][-1][2]
            abstract_list.insert(-1, i)
                
    return positional_index


docs = construct_positional_indexes('')
```

```python
## 1 points

def get_posting_list(word : str):
    
    """ get posting_list of a word
    
        Parameters
        ----------
        word: str
             word we want to check

        Return
        ----------
        dict 
            posting list

    """
    # TODO: Find posting list of a word
    token = clean_data(word)[-1]

    if token not in docs:
        posting_list = {}
    else:
        posting_list = {}
        for occurence in docs[token]:
            paper_id, title, abstract = occurence[0], occurence[1], occurence[2]
            posting_list[paper_id] = {}
            posting_list[paper_id]['title'] = title
            posting_list[paper_id]['abstract'] = abstract

    return posting_list
    

get_posting_list('coronavirus')
```

```python
## 7 points

def add_documnet(document : tuple):
    """Adds a document to positional index

    Parameters
    ----------
    document : str
        Comma separated string containing id,title,abstarct in this exact order
    """
    global stop_words

    # TODO: implement adding a document to the index
    paper_id, title, abstract = document
    if paper_id in mapping.values():
        return 
    doc_id = max(mapping.keys()) + 1
    
    mapping[doc_id] = paper_id

    # Tokenize the title and abstract
    title_tokens =  clean_data(title)
    abstract_tokens = clean_data(abstract)

    # Add title positions to index
    for i, token in enumerate(title_tokens):
        if token in stop_words:
            continue
        if token not in docs:
            docs[token] = []
        if len(docs[token]) == 0 or docs[token][-1][0] != doc_id:
            docs[token].append([doc_id, [len(title)], [len(abstract)]])
        docs[token][-1][1].insert(-1, i)

    # Add abstract positions to index
    for i, token in enumerate(abstract_tokens):
        if token in stop_words:
            continue
        if token not in docs:
            docs[token] = []
        if len(docs[token]) == 0 or docs[token][-1][0] != doc_id:
            docs[token].append([doc_id, [len(title)], [len(abstract)]])
        docs[token][-1][2].insert(-1, i)
        
new_document = ("1eae26fe1ca566f17468080c3aecab1c3f9efb66", 
                "A Deep Learning Framework for Viable Tumor Burden Estimation", 
                "Liver masses have become a common clinical challenge since they require to be defined and accurately categorized as neoplastic or nonneoplastic lesions. Hepatocellular carcinoma (HCC), the most common histologic type of primary liver malignancy, is a global health concern being the fifth most common cancer and the second cause of cancer mortality worldwide. Accurate diagnosis, which in some circumstances requires histopathology results, is necessary for appropriate management. Also, some tumor characteristics help in predicting tumor behavior and patient response to therapy. In this paper, we propose a deep learning framework for the segmentation of whole and viable tumor areas of liver cancer from whole-slide images (WSIs). To this end, we use Fast Segmentation Convolutional Neural Network (Fast-SCNN) as our network. We use the dataset from PAIP 2019 challenge. After data-augmentation on the training subset, we train the network with a multi-term loss function and SWA technique. Our model achieves 0.80 for the median of the Jaccard Index for the task of Viable Tumor Segmentation and 0.77 for the median of Weighted Absolute Accuracy for the task of Viable Tumor Burden Estimation on the whole-slide images of the test subset.")
add_documnet(new_document)
```

```python
## 7 points

def remove_document(document_id : str):
    """removes a document from positional index

    Parameters
    ----------
    document_id : str
        Id of the document
    """

    # TODO: implement removing a document from the index

    for key, value in mapping.items():
        if value == document_id:
            document_id = key
            del mapping[key]
            break

    empty_tokens = []
 
    for token in docs.keys():
        for doc in docs[token]:
            if doc[0] == document_id:
                docs[token].remove(doc)
                if len(docs[token]) == 0:
                    empty_tokens.append(token)
                break
    
    for token in empty_tokens:
        del docs[token]
    return
  
remove_document("1eae26fe1ca566f17468080c3aecab1c3f9efb66")
```

```python
class Utils:
    @staticmethod
    def create_gap_list(num_list):
        gap_list = num_list.copy()
        gap_list = [gap_list[0]]+[gap_list[i]-gap_list[i - 1] for i in range(1,len(gap_list))]
        return gap_list
    
    @staticmethod
    def undo_gap_list(num_list):
        for i in range(1,len(num_list)):
            num_list[i] += num_list[i-1]
        return num_list 
```

```python
class GammaEncoder:

    def __init__(self):
        # Have to make sure each numbers is greater than 2
        self.shift = 2
    
    def gamma_encode(self, numbers):
        bitstring = ""
        for number in numbers:
            number += self.shift
            offset = bin(number)[3:]
            bitstring = bitstring + ("1" *len(offset) + "0" + offset)
        return bitstring
    
    
    def encode_single_doc(self, doc_id, title, abstract):
        title = Utils.create_gap_list(title)
        abstract = Utils.create_gap_list(abstract)

        final_g = self.gamma_encode([doc_id] + [len(title)] + title +  abstract)
        return final_g
    
    def encode(self, docs):
        compressed = {}
        for token in docs:
            prev_doc = docs[token][0][0]
            compressed[token] = [[] * len(docs[token]) for ـ in range(len(docs[token]))]
            title = docs[token][0][1]
            abstract = docs[token][0][2]
            compressed[token][0] = self.encode_single_doc(prev_doc, title, abstract)
            for i in range(1, len(docs[token])):
                doc = docs[token][i][0]
                title = docs[token][i][1]
                abstract = docs[token][i][2]
                compressed[token][i] = self.encode_single_doc(doc - prev_doc, title, abstract)
                prev_doc = doc
        return compressed

class GammaDecoder:

    def __init__(self):
        self.shift = 2

    def gamma_decode(self, bits):
        numbers = []
        while len(bits) > 0:
            index = bits.find('0')
            bits = bits[index + 1:]
            num = int('1' + bits[:index], 2)
            numbers.append(num - self.shift)
            bits = bits[index:]
        return numbers
    
    def decode_single_doc(self, bitstring):
        nums = self.gamma_decode(bitstring)

        doc_id, length = nums[0], nums[1]
        title = Utils.undo_gap_list(nums[2:length + 2])
        abstract = Utils.undo_gap_list(nums[length + 2:])
        
        return [doc_id, title, abstract]
    
    def decode(self, compressed):
        decoded = {}
        for token in compressed:
            decoded[token] = [[] * len(compressed[token]) for ـ in range(len(compressed[token]))]
            decoded[token][0] = self.decode_single_doc(compressed[token][0]) 
            prev_doc = decoded[token][0][0]
            for i in range(1, len(compressed[token])):
                doc = self.decode_single_doc(compressed[token][i])
                decoded[token][i] = [doc[0] + prev_doc, doc[1], doc[2]]
                prev_doc = doc[0] + prev_doc
        return decoded
```

```python
from typing import List


def calculate_precision(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The precision of the predicted results
    """
    precision = 0.0

    for i in range(len(predicted)):
        intersect = set(actual[i]).intersection(set(predicted[i]))
        precision += len(intersect) / len(predicted[i])
    
    precision /= len(predicted)
    
    return precision
```

```python
def calculate_recall(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the recall of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The recall of the predicted results
    """
    recall = 0.0

    for i in range(len(predicted)):
        intersect = set(actual[i]).intersection(set(predicted[i]))
        recall += len(intersect) / len(actual[i])

    recall /= len(predicted)

    return recall
```

```python
def calculate_F1(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the F1 score of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The F1 score of the predicted results    
    """
    f1 = 0.0

    precision = calculate_precision(actual, predicted)
    recall = calculate_recall(actual, predicted)
    f1 = 2 * (precision * recall) / (precision + recall)

    return f1
```

```python
def calculate_MAP(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Average Precision of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The Mean Average Precision of the predicted results
    """
    map = 0.0
    
    for i in range(len(actual)):
        relevant_docs = set(actual[i])
        predicted_docs = predicted[i]
        relevant_count = 0
        precision = 0.0

        for j in range(len(predicted_docs)):
            if predicted_docs[j] in relevant_docs:
                relevant_count += 1
                precision += relevant_count / (j + 1)

        if relevant_count > 0:
            map += precision / relevant_count

    return map / len(actual)
```

```python
def cacluate_NDCG(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Normalized Discounted Cumulative Gain (NDCG) of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The NDCG of the predicted results
    """
    
    ndcg = 0.0
    idcg = 0.0

    p = len(actual[0])
    for i in range(p):
        idcg += (p - i) / math.log2(i + 2)

    for i in range(len(actual)):
        actual_q = actual[i]
        predicted_q = predicted[i]

        dcg = 0.0
        for j in range(len(predicted_q)):
            if predicted_q[j] in actual_q:
                index = actual_q.index(predicted_q[j])
                dcg += (p - index) / math.log2(index + 2)
        ndcg += dcg / idcg

    ndcg /= len(actual)

    return ndcg
```

```python
def cacluate_MRR(actual: List[List[str]], predicted: List[List[str]]) -> float:
    """
    Calculates the Mean Reciprocal Rank of the predicted results

    Parameters
    ----------
    actual : List[List[str]]
        The actual results
    predicted : List[List[str]]
        The predicted results

    Returns
    -------
    float
        The MRR of the predicted results
    """
    MRR = 0.0

    for i in range(len(predicted)):
        for j in range(len(predicted[i])):
            if predicted[i][j] in actual[i]:
                MRR += 1 / (j + 1)
                break
    MRR /= len(predicted)

    return MRR
```
